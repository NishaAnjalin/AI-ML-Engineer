```python
# b) Marks of 10 students: Max, Min, Average
def student_marks_analysis():
    marks = []
    print("Enter marks of 10 students:")
    for i in range(10):
        m = int(input(f"Student {i+1} mark: "))
        marks.append(m)
    print("\n--- Student Marks Analysis ---")
    print("Maximum Mark:", max(marks))
    print("Minimum Mark:", min(marks))
    print("Average Mark:", sum(marks)/len(marks))

# c) Five marks with grades
def student_grades():
    print("\nEnter 5 marks of a student:")
    for i in range(5):
        mark = int(input(f"Mark {i+1}: "))
        if mark > 75:
            grade = "A"
        elif mark >= 65:
            grade = "B"
        elif mark >= 55:
            grade = "C"
        elif mark >= 45:
            grade = "S"
        else:
            grade = "F"
        print(f"Mark: {mark}, Grade: {grade}")

# d) Series of numbers terminated by -999, print sum
def sum_series():
    print("\nEnter numbers (end with -999):")
    total = 0
    while True:
        n = int(input("Enter number: "))
        if n == -999:
            break
        total += n
    print("Sum of numbers entered:", total)

# Run menu for easy testing
def main():
    while True:
        print("\n===== Menu =====")
        print("1. Student Marks Analysis (10 students)")
        print("2. Student Grades (5 marks)")
        print("3. Sum of Series (-999 terminates)")
        print("4. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            student_marks_analysis()
        elif choice == "2":
            student_grades()
        elif choice == "3":
            sum_series()
        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
